The following exercise models a simple system with three binary variables A, B, and C, where A and B are independent and C depends on A and B. The goal is to compute various marginal and conditional probabilities using NumPy operations.

This system models a bell (represented by C) that rings when you throw a coin twice the first through (represented by A) and the second time through (represented by B) and both results are the same. (Either both heads or both tails.)

Consider the coin to be a fair coin. Now model this system using probabilities.

```python
def P_a(a): # p(A)
    p = 0.5
    return a*p + (~a) * (1-p)

def P_b(b): # p(B)
    p = 0.5
    return b*p + (~b) * (1-p)

def P_c_ab(c, a, b): # p(C|A,B)
    # if c == 1 and a == b or c == 0 and a ~=b, then 1.0, otherwise 0.0
    return c == (a == b)

def P_abc(a, b, c): # p(A,B,C)
    return P_a(a) * P_b(b) * P_c_ab(c, a, b)


values = np.asarray([0, 1], dtype=bool)
p = np.array([[[P_abc(a, b, c) for a in values] for b in values] for c in values])

# axis names (that's what the notation "p(A)" means)
A = 0
B = 1
C = 2

# What is the joint probability of A and B when C is unknown?
p_AB = p.sum(axis=C)
# What is the joint probability of A and C when B is unknown?
p_AC = p.sum(axis=B)
# What is the probability of B and C when A is unknown?
p_BC = p.sum(axis=A)

display(
    Markdown(
        "$$p(A,B) = "
        + ltx(p_AB, "{:6.2f}")
        + ", \qquad "
        + "p(A,C) = "
        + ltx(p_AC, "{:6.2f}")
        + ", \qquad "
        + "p(B,C) = "
        + ltx(p_BC, "{:6.2f}")
        + "$$"
    )
)

# What is the probability of A (heads/tails) when B and C are unknown?
p_A = p.sum(axis=(B,C))
# What is the probability of B (heads/tails) when A and C are unknown?
p_B = p.sum(axis=(A,C))
# What is the probability of C (bell ringing/not ringing) when A and B are unknown?
p_C = p.sum(axis=(A,B))
display(
    Markdown(
        "$$p(A) ="
        + ltx(p_A, "{:6.2f}")
        + ", \qquad "
        + "p(B) ="
        + ltx(p_B, "{:6.2f}")
        + ", \qquad "
        + "p(C) ="
        + ltx(p_C, "{:6.2f}")
        + "$$"
    )
)
c = 1
b = 1
a = 1
# What is the joint probability of A and B given that The bell rang?
P_AB_C = p[:, :, c] / p_C
# What is the joint probability of A and C given that B is heads?
P_AC_B = p[:, b, :] / p_B
# What is the joint probability of B and C given that A is heads?
P_BC_A = p[a, :, :] / p_A

display(
    Markdown(
        "$$p(A,B\mid C) = "
        + ltx(P_AB_C, "{:4.2f}")
        + ", \qquad "
        + "p(A,C\mid B) = "
        + ltx(P_AC_B, "{:4.2f}")
        + ", \qquad "
        + "p(B,C\mid A) = "
        + ltx(P_BC_A, "{:4.2f}")
        + "$$"
    )
)
```
